```python
import numpy as np
import pandas as pd
from metpy.units import units
import metpy.calc
```

<font color="forestgreen">Formula and commands for changing temperatures from F to C

```python
def fahr_to_celsius(temp_fahr):
    temp_celsius = (temp_fahr - 32) * 5 / 9
    return temp_celsius
```

<font color="forestgreen">Read in data file

```python
data = pd.read_csv('isusm.txt',header = 0)
```

<font color="forestgreen">Display data

```python
data.head()
```

<font color="forestgreen">set the index as "valid" which is the variable for the date

```python
data = data.set_index('valid')
```

<font color="forestgreen">Convert low temps to celcius and create a new column in the chart

```python
fahr_to_celsius(data["low"])
```

```python
data["low_temp_C"] = fahr_to_celsius(data["low"])
```

```python
data.head()
```

<font color="forestgreen">Repeat for high temperatures

```python
fahr_to_celsius(data["high"])
```

```python
data["high_temp_C"] = fahr_to_celsius(data["high"])
```

```python
data.head()
```

<font color = 'forestgreen'>Define variables for low temp and high temp in celius

```python
low_temp_C = fahr_to_celsius(data["low"])*units.degC
```

```python
high_temp_C = fahr_to_celsius(data["high"])*units.degC
```

<font color="forestgreen">Calculate average temperature, create new column, and display new chart

```python
temp_avg_C = (low_temp_C[:] + high_temp_C[:])/2
data['temp_avg_C']= temp_avg_C
data.head()
```

<font color = "forestgreen">Turn temp average into an array so that sat. vap. pressure can be calculated

```python
temp_avg_C = np.array(temp_avg_C)
```

<font color="forestgreen">Ensure temp avg is in celsius

```python
temp_avg_C = temp_avg_C *units.degC
```

<font color="forestgreen">Calculate sat. vap. press.

```python
metpy.calc.saturation_vapor_pressure(temp_avg_C)
```

<font color="forestgreen">Create new column for sat. vap. press.

```python
data['Saturation Vapor Pressure']= metpy.calc.saturation_vapor_pressure(temp_avg_C)
```
